```python
base = 'https://www.odepa.gob.cl'
link = 'https://www.odepa.gob.cl/contenidos-rubro/boletines-del-rubro/boletin-semanal-de-precios-y-volumenes-de-frutas-y-hortalizas-en-mercados-mayoristas-del-pais'
```

```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import os
```

```python
resp = requests.get(link)
print(resp.status_code)
```

```python
soup = BeautifulSoup(resp.text, "lxml")
elements = soup.select('td > a')
```

```python
to_download = []
for e in elements:
    e_link = e.get('href')
    if 'xls' in e_link:
        to_download.append(e_link)
to_download
```

```python
for i in range(len(to_download)):
    if 'odepa' not in to_download[i]:
        to_download[i] = base + to_download[i]
to_download
```

```python
def download_file(url, save_path):

    if os.path.exists(save_path):
        # The file already exists
        return
    
    response = requests.get(url, stream=True)

    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
    else:
        print(f"Failed to download the file. Status code: {response.status_code} {url}")
```

```python
path = './data/'
for d in to_download:
    fname = path + str(d).split('/')[7]
    download_file(str(d), fname)
    time.sleep(0.3)
```
