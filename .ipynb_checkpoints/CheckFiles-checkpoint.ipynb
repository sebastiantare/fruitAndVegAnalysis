```python
import pandas as pd
import numpy as np
import dateparser
import re
from datetime import datetime
```

```python
def findBetween(datestr, start, end):
    pattern = re.escape(start) + r'(.*?)' + re.escape(end)
    matches = re.findall(pattern, datestr)
    return matches[0]
def extractDates(datestr):
    # Second date
    second_date = dateparser.parse(datestr.split('al ')[1])
    # First date
    first_day = int(findBetween(datestr, 'Semana del ', ' al'))
    first_date = second_date.replace(day=first_day)
    return [first_date, second_date]
```

### Manzanas

```python
df2 = pd.read_excel('./data/Boletin_Semanal_Precios_Mayoristas_20230714.xlsx', sheet_name='Manzana', skiprows=4)
df2.shape
```

```python
df2.columns = np.array(df2.iloc[0])
```

```python
df2.drop(df2[df2['Mercado'].isnull() == True].index, axis=0, inplace=True)
```

```python
df2.drop(df2[df2['Mercado'] == 'Mercado'].index, axis=0, inplace=True)
```

```python
df2 = df2.reset_index(drop=True)
df2
```

```python
df2.rename(columns={'Unidad de\ncomercialización ': 'Unidad'}, inplace=True)
df2[df2['Unidad'].str.contains('\$') == False]
```

## Transform data

```python
price_df = df2[df2['Unidad'].str.contains('\$') == True]
volume_df = df2[df2['Unidad'].str.contains('\$') == False]
price_df = pd.melt(price_df, id_vars=['Variedad', 'Mercado', 'Unidad'], var_name='Dia', value_name='Precio')
volume_df = pd.melt(volume_df, id_vars=['Variedad', 'Mercado', 'Unidad'], var_name='Dia', value_name='Volumen')
volume_df
```

```python
price_df[price_df['Precio'] > 0].groupby('Dia')['Precio'].plot()
```

```python
price_df[price_df['Precio'] > 0].groupby('Variedad')['Precio'].plot()
```

```python
# Precio promedio por dia cuando precio > 0
price_df[price_df['Precio'] > 0].groupby('Dia')['Precio'].mean()
```

```python
# Precio promedio por variedad
price_df.groupby('Variedad')['Precio'].mean()
```

```python
consolidated_df = pd.merge(price_df, volume_df, on=['Variedad', 'Mercado', 'Dia'])

weekdays = {'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 'Friday': 4}
consolidated_df['Weekday_Num'] = consolidated_df['Day'].map(weekdays)

start_date = pd.to_datetime('2023-07-17')

consolidated_df['Date'] = start_date + pd.to_timedelta(consolidated_df['Weekday_Num'], unit='d')

consolidated_df.drop(columns='Weekday_Num', inplace=True)
```
